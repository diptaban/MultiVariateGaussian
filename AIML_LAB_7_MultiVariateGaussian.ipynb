## $N(x|\mu,\sigma)= \frac{1}{\sqrt{2\pi\sigma^2}}exp(-\frac{(x-\mu)^2}{2\sigma^2})$

# **Multivariate Gaussian Distribution**

## $P(x|\mu,\Sigma)=\frac{1}{(2\pi)^{d}||^{1/2}}exp\{-\frac{1}{2}(x-\mu)^T\Sigma^{-1}(x-\mu)\}$



```python

mu = np.array([0.0,0.0])
covariance = [[1,-2],[-2,1]]
x = np.linspace(-3, 3, 101)
y = np.linspace(-3, 3, 101)

d=2
# full coorindate arrays
x1,x2 = np.meshgrid(x, y)
X=np.array([x1,x2])
pos = np.empty(x1.shape + (2,))
print(np.shape(pos))
#print(np.shape(X)[1])
z=[]
N=np.zeros((101,101))
t1 = 1/((2*pi)**d*(np.linalg.det(covariance)**0.5))
x_m=[]
for i in range(np.shape(X)[0]):
  for k in range(np.shape(X)[1]):
    x_m = X[:,i,k]-mu
    N[i,k] = t1*np.exp(-(np.linalg.solve(covariance, x_m).T.dot(x_m)) / 2)
    



#print(N)
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(x1,x2,N,cmap=cm.coolwarm,linewidth=0, antialiased=False)


```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D



def multivariate_gaussian(pos,mu, Sigma):
    """Return the multivariate Gaussian distribution on array pos.

    pos is an array constructed by packing the meshed arrays of variables
    x_1, x_2, x_3, ..., x_k into its _last_ dimension.

    """
    

    n = mu.shape[0]
    Sigma_det = np.linalg.det(Sigma)
    Sigma_inv = np.linalg.inv(Sigma)
    N = np.sqrt((2*np.pi)**n * Sigma_det)
    # This einsum call calculates (x-mu)T.Sigma-1.(x-mu) in a vectorized
    # way across all the input variables.
    fac = np.einsum('...k,kl,...l->...', pos-mu, Sigma_inv, pos-mu)

    return np.exp(-fac / 2) / N

# The distribution on the variables X, Y packed into pos.
mu = np.array([0., 0.])
Sigma = np.array([[ 2 , -1], [-1,  2]])
# Our 2-dimensional distribution will be over variables X and Y
N = 60
X = np.linspace(-3, 3, N)
Y = np.linspace(-3, 3, N)
X, Y = np.meshgrid(X, Y)
# Mean vector and covariance matrix
# Pack X and Y into a single 3-dimensional array
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X
pos[:, :, 1] = Y
Z = multivariate_gaussian( pos,mu, Sigma)
# Create a surface plot and projected filled contour plot under it.
#fig = plt.figure()
#ax = fig.gca(projection='3d')
#ax.plot_surface(X, Y, Z, rstride=3, cstride=3, linewidth=1, antialiased=True,cmap=cm.viridis)
fig,ax=plt.subplots(1,1)
c = ax.contourf(X,Y,Z)
#cset = ax.contourf(X, Y, Z, zdir='z', offset=-0.15, cmap=cm.viridis)
# Adjust the limits, ticks and view angle
#ax.set_zlim(-0.15,0.2)
#ax.set_zticks(np.linspace(0,0.2,5))
#ax.view_init(90, -90)
plt.tight_layout()
plt.show()

```
